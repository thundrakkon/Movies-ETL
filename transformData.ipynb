```python
#TRANSFORM DATA
import pandas as pd
```

```python
# Reading the movies_metadata.csv file into a frame
metadata = pd.read_csv('Resources/movies_metadata.csv')
```

```python
# Getting only the columns we're gonna use into just one dataframe
metadata = metadata[['imdb_id','original_title','title','runtime','budget','revenue',\
                    'popularity','vote_average', 'vote_count']]
metadata
metadata.info()
```

```python
stream_file = "./Resources/MoviesOnStreamingPlatforms_updated.csv"
stream_df = pd.read_csv(stream_file)
stream_df.head()
```

```python
# Create a filtered dataframe from specific columns
stream_cols = ["ID",
               "Title",
               "Year",
               "Age",
               "IMDb",
               "Rotten Tomatoes",
               "Netflix",
               "Hulu",
               "Prime Video",
               "Disney+",
               "Type",
               "Directors",
               "Genres",
               "Country",
               "Language",
               "Runtime"]

stream_transformed= stream_df[stream_cols].copy()

# Rename the column headers
stream_transformed = stream_transformed.rename(columns={"ID": "id",
                                                        "Title": "title",
                                                        "Year": "year",
                                                        "Age": "age",
                                                        "IMDb": "imdb",
                                                        "Rotten Tomatoes": "rotten_tomatoes",
                                                        "Netflix": "netflix",
                                                        "Hulu": "hulu",
                                                        "Prime Video": "prime_video",
                                                        "Disney+": "disney",
                                                        "Type": "type",
                                                        "Directors": "directors",
                                                        "Genres": "genres",
                                                        "Country": "country",
                                                        "Language": "language",
                                                        "Runtime": "runtime"})

# Clean the data by dropping duplicates and setting the index
stream_transformed.drop_duplicates("id", inplace=True)
stream_transformed.set_index("id", inplace=True)

stream_transformed.head()
```

```python
stream_transformed.info()
```

```python
#Merge the two data frames to get the information we want in a clearer form
frame = pd.merge(metadata, stream_transformed, left_on = 'title', right_on='title', how = 'inner')
#Adding a column with what we'll call the 'id_movie', a future primary key
frame['id_movie'] = ''
for i in range(len(frame)):
    frame.iloc[i,-1] = i+1
frame
```

```python
frame.info()
```

```python
frame.to_csv('Resources/frame.csv')
```

```python
frame.revenue = frame.revenue.astype(float)
```

```python
frame.budget = frame.budget.astype(float)
```

```python
frame['profit'] = frame.revenue - frame.budget
```

```python
####################################################################################################
#************* CREATING TABLES *********************************************************************
####################################################################################################

#CREATING THE DIRECTORS TABLE
#first craeating a list with each entry just one name
directors=[]
list_directors = frame.directors.unique()


for item in list_directors:
    try:
        sublist = item.split(',')
        for i in range( len(sublist) ):
            directors.append(sublist[i])
    except:
        directors.append(item)

#turning into an acctual table
directors_table = pd.DataFrame({'name' : directors})
directors_table['id_director'] = ''
directors_table['first_name'] = ''
directors_table['last_name'] = ''
for index, row in directors_table.iterrows():
    row['id_director'] = index + 1
    try:
        row['first_name'] = row['name'].split(' ')[0]
        row['last_name'] = row['name'].split(' ')[-1]
    except:
        row.last_name = row.name
        
directors_table  = directors_table[['id_director','first_name', 'last_name', 'name']]
directors_table
```

```python
directors_table.to_csv("Resources/table_directors.csv", index=False)
```

```python
# CREATING A TABLE FOR GENRES
genres=[]
list_genres = frame.genres.unique()


for item in list_genres:
    try:
        sublist = item.split(',')
        for i in range( len(sublist) ):
            genres.append(sublist[i])
    except:
        genres.append(item)
        
genre_df = pd.DataFrame({'name': genres})
genre_table = pd.DataFrame({'name' : genre_df.name.unique() })
genre_table['id_genre'] = ''

for index, row in genre_table.iterrows():
    row['id_genre'] = index +1
    
genre_table = genre_table[['id_genre', 'name']]
genre_table
```

```python
genre_table.to_csv("Resources/table_genre.csv", index=False)
```

```python
# CREATING A TABLE FOR LANGUAGES
languages=[]
list_lang = frame.language.unique()


for item in list_lang:
    try:
        sublist = item.split(',')
        for i in range( len(sublist) ):
            languages.append(sublist[i])
    except:
        languages.append(item)
        
language_df = pd.DataFrame({'name': languages})
language_table = pd.DataFrame({'name' : language_df.name.unique() })
language_table['id_language'] = ''

for index, row in language_table.iterrows():
    row['id_language'] = index +1
    
language_table = language_table[['id_language', 'name']]
language_table
```

```python
language_table.to_csv("Resources/table_language.csv", index=False)
```

```python
#TABLES ARE DEFINED WE NEED TO CREATE JUNCTION TABLES BECAUSE OS RELATIONS MANY TO MANY
#CRATE A JUNCTION TABLE BETWEEN movie_table AND director_table:
#THIS FUNCTION SHOULD CREATE A JUNCTION TABLE BASED ON THE 'frame'-FRAME WE'VE CREATED
#So it will create junction tables between movie-language, movie-director, movie-genre
#beacuse those are relations many to many.
#Something important is that frame refers to the original frame, table2- to the frame we are 
#comparing to, and in every table2 there exists a 'name'-column.
def junction_table(frame, table2, id_table2, compared):
    #Creating an empty frame
    junction = pd.DataFrame()
    junction['id_movie'] = ''
    junction[id_table2] = ''
    #Initialize the comparision
    for i in range(len(frame)):
        try:
            values = frame.loc[i,compared].split(',')
            for j in range(len(values)):
                for k in range(len(table2)):
                    if values[j] == table2.loc[k,'name']:
                        junction =junction.append({'id_movie': frame.loc[i,'id_movie'], id_table2: table2.loc[k,id_table2]}, ignore_index=True)
                        break;
        except:
            value = frame.loc[i,compared]
            for k in range(len(table2)):
                    if value == table2.loc[k,'name']:
                        junction = junction.append({'id_movie': frame.loc[i,'id_movie'], id_table2: table2.loc[k,id_table2]})
    return junction
```
