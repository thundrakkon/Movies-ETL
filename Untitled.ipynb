```python
import pandas as pd
```

```python
# Reading the movies_metadata.csv file into a frame
metadata = pd.read_csv('Resources/movies_metadata.csv')
```

```python
# Getting only the columns we're gonna use into just one dataframe
metadata = metadata[['imdb_id','original_title','title','runtime','budget','revenue',\
                    'popularity','vote_average', 'vote_count']]
```

```python
metadata
```

```python
metadata.info()
```

```python
stream_file = "./Resources/MoviesOnStreamingPlatforms_updated.csv"
stream_df = pd.read_csv(stream_file)
stream_df.head()
```

```python
# Create a filtered dataframe from specific columns
stream_cols = ["ID",
               "Title",
               "Year",
               "Age",
               "IMDb",
               "Rotten Tomatoes",
               "Netflix",
               "Hulu",
               "Prime Video",
               "Disney+",
               "Type",
               "Directors",
               "Genres",
               "Country",
               "Language",
               "Runtime"]

stream_transformed= stream_df[stream_cols].copy()

# Rename the column headers
stream_transformed = stream_transformed.rename(columns={"ID": "id",
                                                        "Title": "title",
                                                        "Year": "year",
                                                        "Age": "age",
                                                        "IMDb": "imdb",
                                                        "Rotten Tomatoes": "rotten_tomatoes",
                                                        "Netflix": "netflix",
                                                        "Hulu": "hulu",
                                                        "Prime Video": "prime_video",
                                                        "Disney+": "disney",
                                                        "Type": "type",
                                                        "Directors": "directors",
                                                        "Genres": "genres",
                                                        "Country": "country",
                                                        "Language": "language",
                                                        "Runtime": "runtime"})

# Clean the data by dropping duplicates and setting the index
stream_transformed.drop_duplicates("id", inplace=True)
stream_transformed.set_index("id", inplace=True)

stream_transformed.head()
```

```python
stream_transformed.info()
```

```python
frame = pd.merge(metadata, stream_transformed, left_on = 'title', right_on='title', how = 'inner')
```

```python
frame
```

```python
frame.info()
```

```python
frame.to_csv('Resources/frame.csv')
```

```python
frame.revenue = frame.revenue.astype(float)
```

```python
frame.budget = frame.budget.astype(float)
```

```python
frame['profit'] = frame.revenue - frame.budget
```

```python
frame
```

```python
# Creating a table that contains directos
directors=[]
list_directors = frame.directors.unique()


for item in list_directors:
    try:
        sublist = item.split(',')
        for i in range( len(sublist) ):
            directors.append(sublist[i])
    except:
        directors.append(item)
        
directors
```

```python
directors_table = pd.DataFrame({'complete' : directors})
directors_table['id'] = ''
directors_table['name'] = ''
directors_table['last_name'] = ''
directors_table
```

```python
for index, row in directors_table.iterrows():
    row['id'] = index + 1
    try:
        row['name'] = row['complete'].split(' ')[0]
        row['last_name'] = row['complete'].split(' ')[-1]
    except:
        row.last_name = row.complete
        
directors_table
```

```python
# CREATING A TABLE FOR GENRES
genres=[]
list_genres = frame.genres.unique()


for item in list_genres:
    try:
        sublist = item.split(',')
        for i in range( len(sublist) ):
            genres.append(sublist[i])
    except:
        genres.append(item)
        
genre_df = pd.DataFrame({'genre': genres})
genre_table = pd.DataFrame({'genre' : genre_df.genre.unique() })
genre_table['id_genre'] = ''

for index, row in genre_table.iterrows():
    row['id_genre'] = index +1
    
genre_table
```

```python
# CREATING A TABLE FOR LANGUAGES
languages=[]
list_lang = frame.language.unique()


for item in list_lang:
    try:
        sublist = item.split(',')
        for i in range( len(sublist) ):
            languages.append(sublist[i])
    except:
        languages.append(item)
        
languages_df = pd.DataFrame({'language': languages})
languages_table = pd.DataFrame({'genre' : languages_df.language.unique() })
languages_table['id_language'] = ''

for index, row in languages_table.iterrows():
    row['id_language'] = index +1
    
languages_table
```
